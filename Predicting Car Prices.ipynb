```python
import pandas as pd
import numpy as np

cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names=cols)
print('Dataset has',cars.shape[0], 'rows and', cars.shape[1], 'rows')
```

```python
cars.head()
```

I have selected only the columns with continuous values.

```python
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
print('Dataset has',numeric_cars.shape[0], 'rows and', numeric_cars.shape[1], 'rows')
numeric_cars.info()
```

```python
numeric_cars.head()
```

## Data Cleaning

```python
numeric_cars = numeric_cars.replace('?',np.nan)
numeric_cars = numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

Delete all rows which dont have price.

```python
numeric_cars = numeric_cars.dropna(subset=['price'])
```

```python
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
```

```python
numeric_cars.isnull().sum()
```

Normalize all columnns to range from 0 to 1 except the target column.

```python
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
```

```python
numeric_cars.head()
```

## Univariate k-nearest neighbors model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
def knn_train_test(training,target,df):
    np.random.seed(1)
    #Randomize order of rows in df
    random_index = np.random.permutation(df.index)
    rand_df = df.reindex(random_index)
    
    #split the data into two parts
    last_train_row = int(len(rand_df)/2)
    train_df = rand_df.iloc[:last_train_row]
    target_df = rand_df.iloc[last_train_row:]
    
    #Instantiate the KNeighborsRegressor Class
    knn = KNeighborsRegressor()
    
    #Fit the model
    knn.fit(train_df[[training]], train_df[target])
    predictions = knn.predict(target_df[[training]])
    
    #Calculating RMSE
    rmse = mean_squared_error(predictions, target_df[target])** (1/2)
    return rmse
```

```python
training_cols = numeric_cars.columns.drop('price')
rmse_values = {}
for cols in training_cols:
    rmse_value = knn_train_test(cols, 'price', numeric_cars)
    rmse_values[cols] = rmse_value
# Create a Series object from the dictionary so 
# we can easily view the results, sort, etc
rmse_results_series = pd.Series(rmse_values)
rmse_results_series.sort_values()
```

Modified version of knn_train_test() to accept parameter for the k value.

```python
def knn_train_test(training,target,df):
    
    np.random.seed(1)
    #Randomize order of rows in df
    random_index = np.random.permutation(df.index)
    rand_df = df.reindex(random_index)
    
    #split the data into two parts
    last_train_row = int(len(rand_df)/2)
    train_df = rand_df.iloc[:last_train_row]
    target_df = rand_df.iloc[last_train_row:]
    
    k_values = [1,3,5,7,9]
    k_rmses = {}
    #Fit the model
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[training]], train_df[target])
        predictions = knn.predict(target_df[[training]])
    
        #Calculating RMSE
        rmse = mean_squared_error(predictions, target_df[target])** (1/2)
        k_rmses[k] = rmse
    return k_rmses
k_rmse_results = {}

# For each column (minus `price`), train a model, return RMSE value
# and add to the dictionary `rmse_results`.
train_cols = numeric_cars.columns.drop('price')
for col in train_cols:
    rmse_val = knn_train_test(col, 'price', numeric_cars)
    k_rmse_results[col] = rmse_val

k_rmse_results
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k,v in k_rmse_results.items():
    x = [1, 3, 5, 7, 9]
    y = [v[i] for i in x]
    
    plt.plot(x,y)
    plt.xlabel('k value')
    plt.ylabel('RMSE')
```
